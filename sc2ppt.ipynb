```python
from pptx import Presentation
from pptx.util import Inches
import pyscreenshot as ImageGrab

def get_printScreen():
    im = ImageGrab.grab((0,0,1430,900))
    # save image file
    im.save("new_pic.png")
    return "new_pic.png"

def add_slide(prs):
    new_image = get_printScreen()
    blank_slide_layout = prs.slide_layouts[6]
    slide = prs.slides.add_slide(blank_slide_layout)
    left = top = Inches(0)
    pic = slide.shapes.add_picture(new_image, left, top, height= Inches(7.5),width= Inches(10)) 

prs = Presentation()

while True:
    key = input("Take a print screen? (Enter- Yes/ Q- Finish)")
    if key == '':
        add_slide(prs)
    if key == 'q':
        break
    
prs.save('slide.pptx')
```
